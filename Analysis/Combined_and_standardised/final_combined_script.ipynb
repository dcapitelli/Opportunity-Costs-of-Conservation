```python
import rasterstats
import rasterio
from rasterio import features
from rasterio.enums import MergeAlg
from rasterio.plot import show
import pandas 
import geopandas as gpd
import numpy as np
from numpy import int16

version = 'oct_24__'
project_path = "C:/Users/spencerd/Documents/NaturaConnect/Spatial datasets/European countries"
IN_vector_nuts0 = project_path + "/NUTS_regions_missing_countries/reference_vector_feb_24/reference_vector_mar_24_cleaned_NUTS0.shp"
IN_combination_raster = oct_24_OUT_CLC_combination_raster_masked

IN_LULC_new = "C:/Users/spencerd/Documents/NaturaConnect/Spatial datasets/New LULC map/Nov23_EU_landSystem.tif"

OUT_national_mean = project_path + f'/Input_datasets/hrl_combi_step/z_score/{version}national_mean.tif'
OUT_national_stdev = project_path + f'/Input_datasets/hrl_combi_step/z_score/{version}national_stdev.tif'
OUT_z_score = project_path + f'/Input_datasets/hrl_combi_step/z_score/{version}z_score_standardisation.tif'

# read in the reference vector
eu_nuts0 = gpd.read_file(IN_vector_nuts0)
eu_nuts0.columns
# ['NUTS_ID__2', 'LEVL_COD_1', 'CNTR_COD_1', 'NAME_LAT_1', 'NUTS_NAM_1',
#        'layer_2_2_', 'path_2_2_2', 'geometry']
eu_nuts0
```

```python

# read in the crop yield raster
raster_path_1 = IN_combination_raster

# Implementation of lesson into NaturaConnect code
d = rasterstats.zonal_stats(eu_nuts0.geometry, raster_path_1, stats=['mean'])
values = [i['mean'] for i in d]
eu_nuts0['national_mean'] = values

d= rasterstats.zonal_stats(eu_nuts0.geometry, raster_path_1, stats=['std'])
values = [i['std'] for i in d]
eu_nuts0['national_std'] = values


print(eu_nuts0.iloc[:,-6:]) # view the last six columns of the dataframe

print("vector zonal analysis complete")
```

```python

########################################
## Rasterise the mean and standard deviation from the vector ##
########################################

vector = eu_nuts0

burn_value = vector['national_mean']

# Get list of geometries for all features in vector file
geom = [shapes for shapes in vector.geometry]

# Open example raster
raster = rasterio.open(IN_LULC_new)

# create tuples of geometry, value pairs, where value is the attribute value you want to burn
geom_value = ((geom,value) for geom, value in zip(vector.geometry, burn_value))

# Rasterize vector using the shape and transform of the raster
rasterized = features.rasterize(geom_value,
                                out_shape = raster.shape,
                                transform = raster.transform,
                                all_touched = False,
                                fill = -99,   # background value
                                merge_alg = MergeAlg.replace,
                                dtype = float)


# save the rasterized vector out
with rasterio.open(
        OUT_national_mean, "w",
        driver = "GTiff",
        crs = raster.crs,
        transform = raster.transform,
        dtype = rasterio.float32,
        count = 1,
        width = raster.width,
        height = raster.height) as dst:
    dst.write(rasterized, indexes = 1)

print("national mean rasterisation complete")

print("")

print(f"raster available at:  {OUT_national_mean}")



burn_value = vector['national_std']

# Get list of geometries for all features in vector file
geom = [shapes for shapes in vector.geometry]

# Open example raster
raster = rasterio.open(IN_LULC_new)

# create tuples of geometry, value pairs, where value is the attribute value you want to burn
geom_value = ((geom,value) for geom, value in zip(vector.geometry, burn_value))

# Rasterize vector using the shape and transform of the raster
rasterized = features.rasterize(geom_value,
                                out_shape = raster.shape,
                                transform = raster.transform,
                                all_touched = False,
                                fill = -99,   # background value
                                merge_alg = MergeAlg.replace,
                                dtype = float)


# save the rasterized vector out
with rasterio.open(
        OUT_national_stdev, "w",
        driver = "GTiff",
        crs = raster.crs,
        transform = raster.transform,
        dtype = rasterio.float32,
        count = 1,
        width = raster.width,
        height = raster.height) as dst:
    dst.write(rasterized, indexes = 1)

print("national stdev rasterisation complete")

print("")

print(f"raster available at:  {OUT_national_stdev}")


import rasterio
from rasterio.plot import show
import numpy as np
import matplotlib.pyplot as plt

# Open invoerrasters
raster1_pad = OUT_national_mean
raster2_pad = OUT_national_stdev
raster3_pad = IN_combination_raster


with rasterio.open(raster1_pad) as src1, rasterio.open(raster2_pad) as src2, rasterio.open(raster3_pad) as src3:
    # Lees metadata van één invoerraster (ervan uitgaande dat beide rasters dezelfde metadata hebben)
    meta = src1.meta.copy()

    # Werk metadata bij voor het uitvoerraster
    meta.update(driver='GTiff', dtype='float32')

    # Maak uitvoerrasterbestand aan
    uitvoer_pad = OUT_z_score
    with rasterio.open(uitvoer_pad, 'w', **meta) as dst:
        # Voer de vermenigvuldiging uit op de invoerrasters
        mean = src1.read(1)  # Lees eerste band van raster1
        stdev = src2.read(1)  # Lees eerste band van raster2
        raw = src3.read(1)  # Lees eerste band van raster2
        zscore = (raw- mean) / stdev

        # Schrijf het resultaat naar het uitvoerrasterbestand
        dst.write(zscore, 1)  # Schrijf het resultaat naar de eerste band

print(f"Uitvoerraster is succesvol geschreven bij {uitvoer_pad}")
```
